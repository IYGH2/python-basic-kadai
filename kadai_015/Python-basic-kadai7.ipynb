```python
class Human:
  def __init__(self):
    self.name = ""
    self.age = ""

  def set_printinfo(self, name, age):
    self.name = name
    self.age = age

  def printinfo(self):
    print(f"{self.name}さん")
    print(f"{self.age}歳")

Taro = Human()

Taro.set_printinfo("太郎", 20)
Taro.printinfo()
```
