```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}さんは{self.age}歳なので大人です。")
    else:
      print(f"{self.name}さんは{self.age}歳なので大人ではないです。")

humans = [Human("太郎", 20), Human("芳雄", 35), Human("陸", 18), Human("幸子", 40), Human("真由", 19)]

for people in humans:
  people.check_adult()
```
