```python
import random

var = random.randint(1, 50)

print(var)

# if 3の倍数かつ、５の倍数ならTrue = if var % 3 == 0 and var % 5 == 0
# if 3の倍数ならTrue = if var % 3 == 0
# if 5の倍数ならTrue = if var % 5 == 0

if var % 3 == 0 and var % 5 == 0:
  print("FizzBuzz")
elif var % 3 == 0:
  print("Fizz")
elif var % 5 == 0:
  print("Buzz")
else:
  print(var)
```
